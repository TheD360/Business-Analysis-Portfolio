```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas

# print(gas.head())
# print (gas.tail()) 
```

```python
# Plotting the actual graph from the CSV - initial phase

plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)

plt.plot(gas['Year'], gas['Germany']) # you can also use square brackets to access a cloumn or field from the csv file


plt.show()
```

```python
# then we add the title

plt.title('Gas prices overtime (in USD)') # Adding title

plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)
plt.plot(gas['Year'], gas['Germany']) # you can also use square brackets to access a cloumn or field from the csv file


plt.show()
```

```python
# Here we also added the legend, notice the difference.
# Now it is important to understand that all these modificiations can be done in a single block
# I am only breaking it down for understanding in the sequence in which it was done.

plt.title('Gas prices overtime (in USD)')

#plt.figure(figsize=(8,5))

plt.plot(gas.Year, gas.USA, label='United States')
plt.plot(gas.Year, gas.Canada, label='Canada')
plt.plot(gas['Year'], gas['Germany'], label='Germany') # you can also use square brackets to access a cloumn or field from the csv file

plt.legend()

plt.show()
```

```python
gass = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over time (in USD)')

plt.plot(gass.Year, gass.USA, label='United States')
plt.plot(gass.Year, gass.Germany, label='Germany')
plt.plot(gass.Year, gass.Canada, label='United States')

plt.legend()
plt.show()
```
